```python
import cv2
import numpy as np
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import tensorflow as tf

DIR = 'dataset/'
img_size = 80
```

```python
X_train=[]
y_train=[]
for i in tqdm(os.listdir(DIR)):
    fpath = os.path.join(DIR,str(i))
    ffpath = os.path.join(fpath,"train/")
    for img in tqdm(os.listdir(ffpath)):
        path = os.path.join(ffpath,img)
        img = cv2.resize(cv2.imread(path,cv2.IMREAD_GRAYSCALE),(img_size,img_size))
        X_train.append(img)
        y_train.append(int(i)-1)
np.save('X_train.npy',X_train)
np.save('y_train.npy',y_train)
```

```python
X_test=[]
y_test=[]
for i in tqdm(os.listdir(DIR)):
    fpath = os.path.join(DIR,str(i))
    ffpath = os.path.join(fpath,"test/")
    for img in tqdm(os.listdir(ffpath)):
        path = os.path.join(ffpath,img)
        img = cv2.resize(cv2.imread(path,cv2.IMREAD_GRAYSCALE),(img_size,img_size))
        X_test.append(img)
        y_test.append(int(i)-1)
np.save('X_test.npy',X_test)
np.save('y_test.npy',y_test)
```

```python
X_valid=[]
y_valid=[]
for i in tqdm(os.listdir(DIR)):
    fpath = os.path.join(DIR,str(i))
    ffpath = os.path.join(fpath,"valid/")
    for img in tqdm(os.listdir(ffpath)):
        path = os.path.join(ffpath,img)
        img = cv2.resize(cv2.imread(path,cv2.IMREAD_GRAYSCALE),(img_size,img_size))
        X_valid.append(img)
        y_valid.append(int(i)-1)
np.save('X_valid.npy',X_valid)
np.save('y_valid.npy',y_valid)
```
